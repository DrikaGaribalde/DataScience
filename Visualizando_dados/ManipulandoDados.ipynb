```python
dados[["Moeda", "Valor_Anuncio", "Tipo_Anuncio"]]= dados["Valor"].str.split(expand = True)
dados.head()
```

```python
dados[dados["Tipo_Anuncio"].isnull()]["Tipo_Anuncio"].unique() #ficar apenas com o verdadeiro
```

```python
dados_vendas = dados[dados["Tipo_Anuncio"].isnull()]
dados_vendas
```

```python
dados_vendas.info()
```

```python
dados_vendas["Valor_Anuncio"].str.replace(".","").astype(float) 
```

```python
dados_vendas["valor_anuncio_float"] = dados_vendas["Valor_Anuncio"].str.replace(".","").astype(float) 
```

```python
dados_vendas.info()
```

```python
dados_vendas["valor_anuncio_float"].plot.hist(bins = 50)
```

```python
dados_vendas.describe()
```

```python
#pd.set_option("display.precision",2)
pd.set_option("display.float_format", lambda x: '%.2f' %x)
dados_vendas.describe(include="all")
```

```python
dados_vendas[["Rua", "Bairro", "Cidade"]].describe()
```

```python
#gráficos
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(10,8))
ax = sns.histplot(data = dados_vendas, x = "valor_anuncio_float", kde = True) #curva de aproximação kde
ax.set_title("Histograma Valor Imóveis")
plt.xlim(-50, 10000000)
plt.show()
```
